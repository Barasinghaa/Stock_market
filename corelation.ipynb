```python
import numpy as np
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plot
import seaborn
import yfinance as yf
from yahoo_fin.stock_info import get_data
from yahoo_fin.stock_info import get_live_price
```

```python
start=datetime(2023,5,5)
end=datetime(2023,6,9)
```

```python
ticker_list = ["amzn", "aapl", "tsla","goog","meta","nvda"]
historical_datas = []
for ticker in ticker_list:
    r = pd.DataFrame(get_data(ticker,start_date=start,end_date=end))
    df = pd.concat([r])
    df = df.reset_index()
    df = df[['index', 'close', 'ticker']]
    historical_datas.append(df)
historical_datas
```

```python
df=pd.concat(historical_datas)
df
```

```python
newdf=pd.pivot_table(df,values="close",columns="ticker",index=["index"])
newdf
```

```python
corr_df = newdf.corr(method='pearson')
#reset symbol as index (rather than 0-X)
corr_df.head().reset_index()
#del corr_df.index.name
corr_df.head(10)
```

```python
plot.figure(figsize=(13, 8))
seaborn.heatmap(corr_df, annot=True, cmap='RdYlGn')
plot.figure()
```
